```python
import random
import time
import json

# [cite_start]Simula칞칚o do RF01: Exibir n칤veis de contamina칞칚o em tempo real [cite: 113]
# [cite_start]Simula칞칚o do RF04: Emitir alarmes se ultrapassar limites [cite: 116]

LIMITE_CRITICO_PARTICULAS = 100  # Valor hipot칠tico para alarme
ID_SENSOR = "SENSOR_PARTICULAS_01"

def ler_sensor():
    """Simula a leitura de dados do CLP"""
    return random.randint(50, 120)

def monitorar_processo():
    print(f"Iniciando monitoramento do sensor {ID_SENSOR}...")
    print("Pressione o bot칚o 'Parar' do Colab para encerrar.")

    try:
        while True:
            valor_atual = ler_sensor()
            timestamp = time.strftime("%Y-%m-%d %H:%M:%S")

            payload = {
                "timestamp": timestamp,
                "sensor_id": ID_SENSOR,
                "valor": valor_atual,
                "status": "NORMAL"
            }

            # L칩gica de Alarme (RF04)
            if valor_atual > LIMITE_CRITICO_PARTICULAS:
                payload["status"] = "CRITICO - ALARME"
                # O print abaixo simula o alerta visual na IHM
                print(f"游댮 [ALARME] {timestamp} - Valor {valor_atual} excedeu o limite!")
            else:
                print(f"游릭 [INFO] {timestamp} - Leitura: {valor_atual}")

            # Simula intervalo de leitura (RF01 pede tempo real - RNF01 < 1s, ajustado para 2s para visualiza칞칚o)
            time.sleep(2)

    except KeyboardInterrupt:
        print("\nMonitoramento encerrado pelo usu치rio.")

if __name__ == "__main__":
    monitorar_processo()
```
